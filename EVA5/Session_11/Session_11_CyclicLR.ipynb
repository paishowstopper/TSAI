```python
import matplotlib.pyplot as plt
```

```python
x = [*range(0, 30, 5)] # start,stop,step
y = list(map(lambda i: 0 if i%2==0 else 1, x))
```

```python
plt.plot(x, y)
plt.xlabel('Iterations')
plt.ylabel('Learning Rate')
plt.title('Cyclic LR')
plt.show()
```
